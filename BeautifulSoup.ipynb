# BeautifulSoup Examples
This notebook contains multiple examples of using BeautifulSoup for HTML parsing and web scraping.

```python
# Example 1: Basic HTML Parsing
from bs4 import BeautifulSoup

html_doc = """
<html>
  <head><title>My Page</title></head>
  <body>
    <h1>Welcome to my website</h1>
    <p class="description">This is a simple page.</p>
    <a href="https://example.com">Example Link</a>
  </body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Get the title
print(soup.title.string)
# Get the h1 text
print(soup.h1.text)
# Get the paragraph by class
print(soup.find('p', class_='description').text)
# Get the link URL
print(soup.a['href'])
```

```python
# Example 2: Find Multiple Elements
html_doc = """
<ul>
  <li class="fruit">Apple</li>
  <li class="fruit">Banana</li>
  <li class="vegetable">Carrot</li>
</ul>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Find all list items
all_items = soup.find_all('li')
for item in all_items:
    print(item.text)

# Find only fruits
fruits = soup.find_all('li', class_='fruit')
for fruit in fruits:
    print(fruit.text)
```

```python
# Example 3: Using CSS Selectors
html_doc = """
<div id="content">
    <p class="text">Hello World</p>
    <p class="text highlight">Python is fun!</p>
</div>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Select by ID
print(soup.select_one('#content').text)

# Select all paragraphs with class "text"
for p in soup.select('p.text'):
    print(p.text)

# Select specific class
highlighted = soup.select('p.highlight')
print(highlighted[0].text)
```

```python
# Example 4: Extract Attributes
html_doc = """
<a href="https://example.com" title="Example Site">Visit Example</a>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

link = soup.a
print("Link:", link['href'])
print("Title:", link.get('title'))
print("Text:", link.text)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the quotes page
url = "https://quotes.toscrape.com/"

# Send GET request
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all quote blocks
quotes = soup.find_all('div', class_='quote')

# Extract and print quotes
for i, quote in enumerate(quotes, 1):
    text = quote.find('span', class_='text').text
    author = quote.find('small', class_='author').text
    tags = [tag.text for tag in quote.find_all('a', class_='tag')]
    print(f"{i}. {text} — {author}")
    print(f"   Tags: {', '.join(tags)}\n")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Goodreads Quotes page
url = "https://www.goodreads.com/quotes"

# Send GET request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/115.0 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all quote blocks
quote_blocks = soup.find_all('div', class_='quoteText')

# Extract and print quotes
for i, quote in enumerate(quote_blocks, 1):
    # Quotes contain text followed by author
    text = quote.get_text(strip=True).split("―")[0].strip()
    author = quote.find('span', class_='authorOrTitle').text.strip()
    print(f"{i}. \"{text}\" — {author}\n")
```
