# Built In Iterators in python

### filter

```python
nums=[1,2,3,4]
```

```python
evens=filter(lambda x:x%2==0,nums)
print(list(evens))
```

```python
odds=filter(lambda x:x%2!=0,nums)
print(list(odds))
```

### Map

```python
squared=map(lambda a: a**2, nums)
print(list(squared))
```

### Zip

```python
a=[1,2,3]
b=['a','b']
c=[55,77]
z=zip(a,b,c)
print(list(z))
```

## Library itertools

### count

```python
from itertools import count

for i in count(10):
    if i > 15:
        break
    print(i)
```

### cycle

```python
from itertools import cycle
c=0
for i in cycle(['Red','Yellow','Green']):
    if(c>10):
        break
    c+=1
    print(i)
```

### chain

```python
from itertools import chain

a=[1,2]
b=[3,4]
print(list(chain(a,b)))
```

### permutations

```python
from itertools import permutations

print(list(permutations([1,2,3])))
```

### combinations

```python
from itertools import combinations

print(list(combinations([1,2,3],2)))
```

### groupby

```python
from itertools import groupby

data=[1,1,2,2,3]#make sure data is sorted

for key,group in groupby(data):
    print(key,list(group))
```

### Generator Expressions (Built-in Syntax)

```python
gen= (x*x for x in range(5))
print(type(gen))
print(list(gen))
```
