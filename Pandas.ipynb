## Data Frames

```python
data={'state':['Ohio','Ohio','Ohio','Delhi','Delhi'],
      'year':[2005,2006,2007,2008,2009],
      'pop':[1.5,1.7,3.6,2.4,2.9]}

print(type(data))
print(data)
```

```python
print(data.keys())
```

```python
df=DataFrame(data)
print(df)
```

```python
df2=DataFrame(data,columns=['year','state','pop','debt'])
print(df2)
```

```python
val=Series([-1.2,-1.5,-1.7],index=[0,1,2])
df2['debt']=val
print(df2)
```
